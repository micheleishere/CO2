```python
from time import time
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
```

```python
# get face image data
faces = fetch_lfw_people(min_faces_per_person=70, resize=0.4)

X, y = faces.data, faces.target
_, h, w = faces.images.shape
print("image size:", h, w)
X.shape
```

```python
# derive train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Train a SVM classification model

clf = SVC(kernel="rbf", C=1, class_weight="balanced")
clf = clf.fit(X_train, y_train)
```

```python
# Predicting people's names on the test set
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)

for i, name in enumerate(faces.target_names):
    print("Accuracy for %17s: %.3f" % (name,accuracy_score(y_pred==i,y_test==i)))
```

```python
import random 

def plot_gallery(images):
    """Helper function to plot a gallery of portraits"""
    n_row, n_col = 3,4
    
    names  = [n.rsplit(" ", 1)[-1] for n in faces.target_names]
    titles = ["predicted: %s\ntrue: %s" % (names[y_pred[i]], names[y_test[i]]) for i in range(len(y_pred))]

    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=0.01, right=0.99, top=0.90, hspace=0.35)
    j = 0
    for i in random.sample(range(len(images)),n_row * n_col):
        plt.subplot(n_row, n_col, j + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())
        j += 1

plot_gallery(X_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
import pandas as pd

#Cs = np.arange(0.1,10.1,0.1)
Cs = [0.2,0.5,1,10,100]
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
df = pd.DataFrame(columns=['accuracy','dataset','C','kernel'])

for C in Cs:
    for kernel in kernels:
        clf = SVC(kernel=kernel, C=C, class_weight="balanced")
        clf = clf.fit(X_train, y_train)
        df.loc[len(df)] = [accuracy_score(y_train, clf.predict(X_train)), \
                           'train',C,kernel]
        df.loc[len(df)] = [accuracy_score(y_test, clf.predict(X_test)), \
                           'test',C,kernel]
```

```python
import seaborn as sns
sns.lineplot(df[df['kernel']=='rbf'], x='C',y='accuracy',hue='dataset')
```

```python
sns.relplot(
    data=df, x="C", y="accuracy",
    col="kernel", hue='dataset',
    kind="line"
)
```
