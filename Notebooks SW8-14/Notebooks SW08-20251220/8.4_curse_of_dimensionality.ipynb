```python
import numpy as np
from pylab import plot
```

```python
# create random points in hypercube
def sample(d=2, N=1000):
    return [[np.random.uniform(0., 1.) for i in range(d)] for _ in range(N)]
```

```python
# evaluation function:
# count how many points are outliers
def corner_count(points):
    return np.mean([any([(d < .01 or d > .99) for d in p]) for p in points])
```

```python
# plot the fraction of points near the corners of the hypercube (blue) as the dimension increases and comparing it to a theoretical curve (orange)
Ds=range(1,20)
x=[corner_count(sample(d)) for d in Ds]
x=np.array(x)
print(x.shape)
plot(Ds, [corner_count(sample(d)) for d in Ds])
plot(Ds, [1 - 0.98**d for d in Ds])
```

```python
# run everything
Ds=range(1,200)
plot(Ds, [corner_count(sample(d)) for d in Ds])
plot(Ds, [1 - 0.98**d for d in Ds])

# illustrate the curse of dimensionality by showing how the fraction of points near the corners of the hypercube increases rapidly with dimension
```
