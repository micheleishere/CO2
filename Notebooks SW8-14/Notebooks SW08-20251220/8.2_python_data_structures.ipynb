```python
# Assign a string to a variable
a = "hello"
print(a) 
```

```python
# access elements of string by index/square brackets
print(a[1])
```

```python
# Strings are arrays and can be looped through
for x in "deep":
  print(x)
```

```python
# Check whether a certain phrase or character is present in a string  
text = "Machine learning is cool!"
print("cool" in text)
```

```python
# The strip() method removes any whitespace from the beginning or the end
a = "   Hello, World!    "
print(a)
print(a.strip()) 
```

```python
# The split() method splits the string into substrings if it finds instances of the separator
# and creates a list.
a = "Hello, World!"
print(a)
print(a.split(","))
```

## Lists ##

```python
list_names = ['tom', 'jerry', 'spike', 'tyke']
print(list_names)
```

```python
# Access an element by index/square brackets
print(list_names[1])
```

```python
# Negative indexing reverses the indexing
print(list_names[-1])
```

```python
# Access range of values by specifying range of indexes 
print(list_names[1:3])
```

```python
# Add an item to the end of the list
list_names.append("mike")
print(list_names)
```

```python
# You can loop through the list items by using a for loop:
for x in list_names:
  print(x)
```

```python
# Create a new list based on the values of an existing list by list comprehension:    
new_list = []
for x in list_names:
  if "t" in x:
    new_list.append(x)
print(new_list)
```

```python
# List comprehension offers a short-cut:
new_list = [x for x in list_names if "t" in x]
print(new_list) 
```

## Dictionaries ##

```python
# Dictionaries are written with curly brackets, and have keys and values:
dictionary = {
  "course": "ML&PR",
  "location": "Olten",
  "year": 2021
}
print(dictionary)    
```

```python
# Adding an item to the dictionary is done by using a new index key and assigning a value to it
dictionary["language"] = "English"
print(dictionary) 
print(dictionary["language"])
```

```python
# Get a list of the keys or the values
dictionary.keys() 
dictionary.values() 
```

```python
# If you try to extract a value using a non-existent key, you get a key-error
dictionary['teacher']
```

```python
# Get a list of the key:value pairs, as tuples in a list
print(dictionary.items()) 
```

```python
# create dictionary from two lists
keys = ["a", "b"]
values = [1, 2]
key_value_pairs = zip(keys, values)
# The zip() function takes an iterable—such as a list, tuple, set, or dictionary—as an argument. 
# The function will generate a list of tuples that contain elements from each iterable.
```

```python
# The dict() constructor builds dictionaries directly from sequences of key-value pairs
dictionary = dict(key_value_pairs)
print(dictionary)
```
