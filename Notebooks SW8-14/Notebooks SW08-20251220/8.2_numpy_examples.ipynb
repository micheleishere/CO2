```python
# calculate the element-wise product of two matrices
A = np.array([[1, 1],[0, 1]])
B = np.array([[2, 0],[3, 4]])
A * B 
```

```python
# calculate the matrix product
A @ B 
```

```python
# a new matrix
b = np.arange(12).reshape(3, 4)
b
```

```python
# sum of each column
b.sum(axis=0) 
```

```python
# min of each row
b.min(axis=1)
```
