```python
import numpy as np
from pylab import plot
```

```python
# create random points in hypercube
def sample(d=2, N=1000):
    return [[np.random.uniform(0., 1.) for i in range(d)] for _ in range(N)]
```

```python
# evaluation function:
# count how many points are outliers
def corner_count(points):
    return np.mean([any([(d < .01 or d > .99) for d in p]) for p in points])
```

```python
# plot the fraction of points near the corners of the hypercube (blue) as the dimension increases and comparing it to a theoretical curve (orange)
Ds=range(1,20)
x=[corner_count(sample(d)) for d in Ds]
x=np.array(x)
print(x.shape)
plot(Ds, [corner_count(sample(d)) for d in Ds])
plot(Ds, [1 - 0.98**d for d in Ds])
```

```python
# run everything
Ds=range(1,200)
plot(Ds, [corner_count(sample(d)) for d in Ds])
plot(Ds, [1 - 0.98**d for d in Ds])

# illustrate the curse of dimensionality by showing how the fraction of points near the corners of the hypercube increases rapidly with dimension
```

```python
# --->>> Your Turn <<<---
# can you explain why the theoretical curve is 1 - 0.98**d?

# the volume of the hypercube within 
# (i.e. without the 0.01 margin on each side) is 0.98^d
# thus the volume outside is 1 - 0.98^d
```

```python
# --->>> Additional Comment <<<---
# If we calculate the average distance of two random points in a d-dimensional unit hypercube
# and plot it as a function of d, we see that the average distance increases with d.
# This is another illustration of the curse of dimensionality.

Ds=range(1,200)
avg_distances = []
for d in Ds:
    data = sample(d)
    from scipy.spatial.distance import pdist
    distances = pdist(data)
    avg_distances.append(np.mean(distances))
    # the average distance increases with d

plt.plot(Ds, avg_distances)
plt.xlabel('Dimension d')
plt.ylabel('Average distance between two random points')
plt.title('Average distance between two random points in a d-dimensional unit hypercube')
plt.show()
```
