# Linear Regression

```python
# data
X, y = [1,2,3,4], [6,5,7,10]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y);
```

```python
# fit() from sklearn uses the analytical solution
import numpy as np
from sklearn.linear_model import LinearRegression
X = np.asarray(X).reshape(-1, 1)
y = np.asarray(y).reshape(-1, 1)
model = LinearRegression().fit(X, y)
```

```python
model.intercept_, model.coef_
```

```python
X_lin = np.linspace(0,6,100)
y_lin = 3.5 + X_lin*1.4
plt.scatter(X,y, color='red')
plt.plot(X_lin,y_lin);
```

```python
# fit() by scanning through the parameter 
# to make it simpel we only vary the intercept parameter
def fit2(X,y):
    lbeta0 = []
    lloss = []
    for beta0 in np.linspace(-10,10,500):
        y_hat = beta0 + (1.4)*X
        loss = ((y - y_hat)**2).sum()
        lloss.append(loss)
        lbeta0.append(beta0)
    return lbeta0[np.argmin(lloss)]
fit2(X,y)
```

```python
# plot the loss against beta0
lbeta0 = []
lloss = []
for beta0 in np.linspace(-10,10,500):
    y_hat = beta0 + (1.4)*X
    loss = ((y - y_hat)**2).sum()
    lloss.append(loss)
    lbeta0.append(beta0)
plt.plot(lbeta0,lloss);
```

```python
# fit() using Gradient descent
n_epoch = 1000
learning_rate = 0.01
def fit3(X,y):
    beta0 = 0
    beta1 = 0
    for epoch in range(n_epoch):
        y_hat = beta0 + beta1*X
        dbeta0 = -(y - y_hat).sum()
        dbeta1 = -((y - y_hat)*X).sum()
        beta0 = beta0 - learning_rate * dbeta0    
        beta1 = beta1 - learning_rate * dbeta1  
    return beta0, beta1
fit3(X,y)
```
